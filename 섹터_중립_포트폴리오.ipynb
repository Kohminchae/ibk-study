```python
from numpy import NaN
stocks = fdr.StockListing('KOSPI')
stocks
```

```python
stocks.dropna(axis=0, inplace = True)
stocks
```

```python
s = stocks['Symbol'].head(100)
```

```python
df = []
index = []
for i in s:
  try:
    close = fdr.DataReader(i,'2021-05-13','2022-05-13')['Close']
  except:
    continue
  df.append(close)
  index.append(i)
```

```python
d = pd.concat(df,axis=1)
d.columns = index
```

```python
d = d.pct_change()
```

```python
d = d.shift(-1)
```

```python
d
```

```python
k = (1+d).cumprod() - 1
```

```python
k = k.drop(['2022-05-13'])
```

```python
k
```

```python
momentum = k.loc['2022-05-12']
```

```python
df_m = pd.DataFrame({'momentum':momentum, 'sector':momentum})
df_m.index = k.columns
```

```python
df = fdr.StockListing('KOSPI')
```

```python
for i in df_m.index:
  s = df[df['Symbol'] == i]['Sector']
  if s.iloc[0] == NaN:
    df.drop(i,inplace = True)
    continue
  df_m['sector'][i] = s.iloc[0]
```

```python
df_m
```

```python
top = df_m.sort_values('momentum',ascending=False).head(50)
top
```

```python
top_group_sector = top.groupby('sector')
top_group_sector.describe()
```

```python
momentum_count = top_group_sector.count()['momentum'].sort_values(ascending=False).head(10)
momentum_count
```

```python
for i in momentum_count.index:
  s = top[top['sector'] == i]
  std = s.std()['momentum']
  mean = s.mean()['momentum']
  nor = (top_group_sector.get_group(i)['momentum'] - mean )/std

  top_group_sector.get_group(i)['momentum'] = nor
  for j in range(len(top)):
    if top['sector'].iloc[j] == s['sector'][0]:
      top['momentum'].iloc[j] = (top['momentum'].iloc[j]-mean)/std
```

```python
top_g = top.groupby('sector')
top_g.describe()
```
